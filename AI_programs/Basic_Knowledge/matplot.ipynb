绘制折线图

```python
import matplotlib.pyplot as plt

# 数据准备
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
# 绘制数据
plt.plot(x, y)
# 添加标题和轴标签
plt.title("example")
plt.xlabel("X")
plt.ylabel("Y")
# 显示图像
plt.show()
```

使用NumPy生成一组符合正态分布的随机数，然后用Matplotlib来绘制这些数据的直方图，

```python
import numpy as np
import matplotlib.pyplot as plt
# 使用NumPy生成1000个符合标准正态分布（均值为0，标准差为1）的随机数
data = np.random.randn(1000)
# 创建图形
plt.figure(figsize=(8, 6))
# 绘制直方图
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
# 添加标题和轴标签
plt.title("Random Numbers Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
# 显示图像
plt.show()
```
