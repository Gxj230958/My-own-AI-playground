# 神经网络手动推导作业

## 网络结构
- 输入层：2个神经元（x1、x2）
- 隐藏层：4个神经元
- 激活函数：Sigmoid
- 输出层：1个神经元
- 损失函数：均方误差(MSE): (y'-y)^2

## 样本数据
- 特征：x1 = 0.5, x2 = 1.0
- 真实值：y = 0.8
- 学习率：0.1

## 神经网络结构图

```
输入层          隐藏层                输出层
            ┌─── h1 ───┐
            │          │
x1 ─────────┼─── h2 ───┼───── y'
            │          │
x2 ─────────┼─── h3 ───┼
            │          │
            └─── h4 ───┘
```

## 初始化参数

### 输入层到隐藏层的权重和偏置

输入层到隐藏层的权重矩阵W^(1)：

```
W^(1) = | 0.1  0.2 |
        | 0.3  0.1 |
        | 0.2  0.3 |
        | 0.1  0.4 |
```

隐藏层的偏置向量b^(1)：

```
b^(1) = | 0.1 |
        | 0.2 |
        | 0.1 |
        | 0.3 |
```

### 隐藏层到输出层的权重和偏置

隐藏层到输出层的权重向量W^(2)：

```
W^(2) = | 0.2  0.3  0.1  0.4 |
```

输出层的偏置b^(2)：

```
b^(2) = 0.1
```

## 正向传播（前馈计算）

### 步骤1：计算隐藏层的输入和输出

隐藏层的输入z^(1)：

z1^(1) = w11^(1) · x1 + w12^(1) · x2 + b1^(1) = 0.1 · 0.5 + 0.2 · 1.0 + 0.1 = 0.05 + 0.2 + 0.1 = 0.3500

z2^(1) = w21^(1) · x1 + w22^(1) · x2 + b2^(1) = 0.3 · 0.5 + 0.1 · 1.0 + 0.2 = 0.15 + 0.1 + 0.2 = 0.4500

z3^(1) = w31^(1) · x1 + w32^(1) · x2 + b3^(1) = 0.2 · 0.5 + 0.3 · 1.0 + 0.1 = 0.1 + 0.3 + 0.1 = 0.5000

z4^(1) = w41^(1) · x1 + w42^(1) · x2 + b4^(1) = 0.1 · 0.5 + 0.4 · 1.0 + 0.3 = 0.05 + 0.4 + 0.3 = 0.7500

应用Sigmoid激活函数σ(z) = 1/(1 + e^(-z))计算隐藏层的输出h：

h1 = σ(z1^(1)) = 1/(1 + e^(-0.35)) = 1/(1 + 0.7047) = 1/1.7047 = 0.5866

h2 = σ(z2^(1)) = 1/(1 + e^(-0.45)) = 1/(1 + 0.6376) = 1/1.6376 = 0.6106

h3 = σ(z3^(1)) = 1/(1 + e^(-0.5)) = 1/(1 + 0.6065) = 1/1.6065 = 0.6225

h4 = σ(z4^(1)) = 1/(1 + e^(-0.75)) = 1/(1 + 0.4724) = 1/1.4724 = 0.6791

### 步骤2：计算输出层的输入和输出

输出层的输入z^(2)：

z^(2) = w11^(2) · h1 + w12^(2) · h2 + w13^(2) · h3 + w14^(2) · h4 + b^(2)
z^(2) = 0.2 · 0.5866 + 0.3 · 0.6106 + 0.1 · 0.6225 + 0.4 · 0.6791 + 0.1
z^(2) = 0.1173 + 0.1832 + 0.0623 + 0.2716 + 0.1 = 0.7344

应用Sigmoid激活函数计算输出y'：

y' = σ(z^(2)) = 1/(1 + e^(-0.7344)) = 1/(1 + 0.4798) = 1/1.4798 = 0.6758

### 步骤3：计算损失

使用均方误差(MSE)损失函数：

L = (y' - y)^2 = (0.6758 - 0.8)^2 = (-0.1242)^2 = 0.0154

## 反向传播

### 步骤1：计算输出层的梯度

首先计算损失函数对输出的梯度：

∂L/∂y' = 2(y' - y) = 2 · (0.6758 - 0.8) = 2 · (-0.1242) = -0.2484

计算输出对输出层输入的梯度（Sigmoid导数）：

∂y'/∂z^(2) = y' · (1 - y') = 0.6758 · (1 - 0.6758) = 0.6758 · 0.3242 = 0.2191

计算损失函数对输出层输入的梯度（链式法则）：

∂L/∂z^(2) = ∂L/∂y' · ∂y'/∂z^(2) = -0.2484 · 0.2191 = -0.0544

计算损失函数对输出层权重的梯度：

∂L/∂w11^(2) = ∂L/∂z^(2) · h1 = -0.0544 · 0.5866 = -0.0319

∂L/∂w12^(2) = ∂L/∂z^(2) · h2 = -0.0544 · 0.6106 = -0.0332

∂L/∂w13^(2) = ∂L/∂z^(2) · h3 = -0.0544 · 0.6225 = -0.0339

∂L/∂w14^(2) = ∂L/∂z^(2) · h4 = -0.0544 · 0.6791 = -0.0369

计算损失函数对输出层偏置的梯度：

∂L/∂b^(2) = ∂L/∂z^(2) = -0.0544

### 步骤2：计算隐藏层的梯度

计算损失函数对隐藏层输出的梯度：

∂L/∂h1 = ∂L/∂z^(2) · w11^(2) = -0.0544 · 0.2 = -0.0109

∂L/∂h2 = ∂L/∂z^(2) · w12^(2) = -0.0544 · 0.3 = -0.0163

∂L/∂h3 = ∂L/∂z^(2) · w13^(2) = -0.0544 · 0.1 = -0.0054

∂L/∂h4 = ∂L/∂z^(2) · w14^(2) = -0.0544 · 0.4 = -0.0218

计算隐藏层输出对隐藏层输入的梯度（Sigmoid导数）：

∂h1/∂z1^(1) = h1 · (1 - h1) = 0.5866 · (1 - 0.5866) = 0.5866 · 0.4134 = 0.2425

∂h2/∂z2^(1) = h2 · (1 - h2) = 0.6106 · (1 - 0.6106) = 0.6106 · 0.3894 = 0.2378

∂h3/∂z3^(1) = h3 · (1 - h3) = 0.6225 · (1 - 0.6225) = 0.6225 · 0.3775 = 0.2350

∂h4/∂z4^(1) = h4 · (1 - h4) = 0.6791 · (1 - 0.6791) = 0.6791 · 0.3209 = 0.2179

计算损失函数对隐藏层输入的梯度（链式法则）：

∂L/∂z1^(1) = ∂L/∂h1 · ∂h1/∂z1^(1) = -0.0109 · 0.2425 = -0.0026

∂L/∂z2^(1) = ∂L/∂h2 · ∂h2/∂z2^(1) = -0.0163 · 0.2378 = -0.0039

∂L/∂z3^(1) = ∂L/∂h3 · ∂h3/∂z3^(1) = -0.0054 · 0.2350 = -0.0013

∂L/∂z4^(1) = ∂L/∂h4 · ∂h4/∂z4^(1) = -0.0218 · 0.2179 = -0.0047

计算损失函数对隐藏层权重的梯度：

∂L/∂w11^(1) = ∂L/∂z1^(1) · x1 = -0.0026 · 0.5 = -0.0013

∂L/∂w12^(1) = ∂L/∂z1^(1) · x2 = -0.0026 · 1.0 = -0.0026

∂L/∂w21^(1) = ∂L/∂z2^(1) · x1 = -0.0039 · 0.5 = -0.0020

∂L/∂w22^(1) = ∂L/∂z2^(1) · x2 = -0.0039 · 1.0 = -0.0039

∂L/∂w31^(1) = ∂L/∂z3^(1) · x1 = -0.0013 · 0.5 = -0.0007

∂L/∂w32^(1) = ∂L/∂z3^(1) · x2 = -0.0013 · 1.0 = -0.0013

∂L/∂w41^(1) = ∂L/∂z4^(1) · x1 = -0.0047 · 0.5 = -0.0024

∂L/∂w42^(1) = ∂L/∂z4^(1) · x2 = -0.0047 · 1.0 = -0.0047

计算损失函数对隐藏层偏置的梯度：
