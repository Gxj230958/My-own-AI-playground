请构造一个神经网络，要求如下：

    输入层：2 个神经元（x1、x2）；
    隐藏层：4 个神经元
    激活函数：Sigmoid
    输出层：1 个神经元
   
    损失函数：均方误差 (MSE):(1/1)*(y'-y)^2
其中需要计算的样本有两个特征：x1 = 0.5, x2 = 1.0 ；样本的真实值：y = 0.8；学习率：0.1
请手动推导其正向前馈计算和反向传播过程（2步即可）。

请手写计算，拍照保存为PDF文档并上传elearning

```python
import torch
import torch.nn as nn
import torch.optim as optim

# 定义数据集
X = torch.tensor([[0.5, 0.1]])
y = torch.tensor([[0.8]])

# 设置超参数
learning_rate = 0.1
num_epochs = 2

# 定义模型
class SimpleNN(nn.Module):
    def __init__(self, input_size=2,hidden_size=4, output_size=1):#构造函数
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size,dtype=torch.float32) #输入到隐层
        self.sigmoid = nn.Sigmoid()#激活层
        self.fc2 = nn.Linear(hidden_size, output_size,dtype=torch.float32)#隐层到输出      
        # 手动将权重和偏置初始化为零
        nn.init.constant_(self.fc1.weight, 0)
        nn.init.constant_(self.fc1.bias, 0)   
        nn.init.constant_(self.fc2.weight, 0)
        nn.init.constant_(self.fc2.bias, 0)   
           
    def forward(self, x):#前向计算
        x=self.fc1(x)
        x=self.sigmoid(x)
        out=self.fc2(x)
        return out

# 初始化模型、损失函数和优化器
model = SimpleNN(input_size=2,hidden_size=4,output_size=1)
criterion = nn.MSELoss()#损失函数的输入是模型输出和参考答案
optimizer = optim.SGD(model.parameters(), lr=learning_rate)#优化器和模型参数绑定

# 训练模型
for epoch in range(num_epochs):
    # 前向传播
    y_pred = model(X)
  
    # 计算损失
    loss = criterion(y_pred, y)
    
    # 反向传播
    optimizer.zero_grad()
    loss.backward()
    
    # 更新权重和偏置
    optimizer.step()
    
    # 打印每个epoch的损失
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.6f}, y_pred:{y_pred}')

    # 输出模型
    for name, param in model.named_parameters():
        if param.requires_grad:
            print(f"{name}: {param.data}:{param.grad}")
```
