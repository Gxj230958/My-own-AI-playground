PyTorch 是一个基于 Torch 库的开源机器学习框架，由 Facebook 的 AI 研究团队开发并维护。它广泛应用于深度学习研究，因其灵活性和动态计算图而备受青睐。以下是PyTorch核心组件：\
torch.Tensor\
张量（Tensor）：类似于 NumPy 的 ndarray，并且支持自动求导和 GPU 加速。\
操作：提供各种数学运算、矩阵操作等功能。\
<br>
nn.Module
模块（Module）：用于定义神经网络层和整个模型。
继承：通过继承 nn.Module 类来创建自定义层和模型。\
<br>
autograd
自动求导：用于计算梯度，支持链式法则。\
<br>
optim
优化器（Optimizer）：提供多种优化算法，如 SGD、Adam、RMSprop 等。\
简单易用：通过定义优化器对象并调用 step() 方法来更新参数。\
<br>
DataLoader\
数据加载器：用于加载和预处理数据集，支持批量处理、打乱数据等。\
方便集成：与 torchvision 等库配合使用，简化数据准备过程。

张量属性

```python
import torch
import numpy as np

# 创建一个基于现有数据的张量
data = np.array([[1, 2, 3], [4, 5, 6]])
x = torch.tensor(data)

# 获取张量的各种属性
shape = x.shape
dtype = x.dtype
device = x.device
dimensions = x.dim()
num_elements = x.numel()
is_cpu = x.is_cpu
is_cuda = x.is_cuda

# 打印属性
print("Tensor x:")
print(x)
print("\nShape of Tensor x:", shape)
print("Dtype of Tensor x:", dtype)
print("Device of Tensor x:", device)
print("Dimensions of Tensor x:", dimensions)
print("Number of Elements in Tensor x:", num_elements)
print("Is Tensor on CPU:", is_cpu)
print("Is Tensor on CUDA:", is_cuda)
```

```python
a = torch.tensor([4,4])
b = torch.tensor([1,1])
w = torch.tensor([[1,1],[2,2],[3,3]])
m = torch.tensor([[1,2],[3,4]])
x1=torch.matmul(a,b)#向量点积
x2=torch.matmul(w,b)#矩阵向量乘法
x3=torch.mm(w,m)#矩阵乘法
x4=torch.matmul(w,m)#高维张量乘法
x1,x2,x3,x4
```

张量拼接操作

```python
import torch
import numpy as np

# 创建一个形状为 (2, 2, 4) 的三维张量
tensor_3d_0 = torch.zeros(16).reshape(2, 2, 4)
tensor_3d_1 = torch.ones(16).reshape(2, 2, 4)
# 使用 torch.cat 将两个三维张量拼接成四维张量
tensor_4d_cat = torch.cat((tensor_3d_0.unsqueeze(0), tensor_3d_1.unsqueeze(0)), dim=0)
tensor_3d_0, tensor_3d_1,tensor_4d_cat
```

向量化

```python
import torch # 导入PyTorch库
import time #导入时间库
# 检查是否有可用的GPU，并选择合适的设备
torch.manual_seed(0)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#device = torch.device("mps" if torch.mps.is_available() else "cpu")
print(f"Using device: {device}")
# 创建两个一百万维度的随机张量a和b
a = torch.rand(1000000, device=device,dtype=torch.float32)#device=device,
b = torch.rand(1000000,device=device,dtype=torch.float32)
#c = torch.zeros(1,device=device,dtype=torch.float64)

# 非向量化版本
c = 0
start_time = time.time()
an=a.numpy()
bn=b.numpy()
for i in range(len(a)):
    c += a[i] * b[i]
end_time = time.time()
print("%.9f"%c)
print("For循环方法运行时间: %.6f ms" % ((end_time - start_time) * 1000))

# 向量化版本
start_time = time.time() # 记录当前时间
c = torch.dot(a, b)
end_time = time.time()
print("%.9f"%c) # 使用item()方法将单元素张量转换为Python标量
print("向量化方法运行时间: %.6f ms" % ((end_time - start_time) * 1000))
```
